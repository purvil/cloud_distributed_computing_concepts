* Representational state transfer in Applicaiton Programming Interface

* REST is data architecture and design method, Receives standard method called verbs and return structured data in JSON or XML format. (resource)
* In old days we request the server which returns HTML, CSS all the times. Each page needs to generated by developer or content management system before it is downloaded.
* Now think about web application, First time we download entire HTML, CSS, JavaScript. Application sends URI request to know next state of the application. Response is used to modify state of web application. This transfered info is data object not entire file.
* All different web application use same resource, mobile app and web app both use same data as REST resource.
* URI : Sequence of characters that identifies abstract or physical resources
* URL: Subset of URI that identifies the resource and explain how to access that resource by providing method like http or ftp

### 6 Constraints of REST
1. Client-server architecture
    - Client manages UI concerns and server manages data storage concerns. One REST server is handling multiple client without knowing what they look like or what they are doing.
    - Separation of content and its presentation
2. Statelessness
    - No client context or info can be saved on the server between the request.
3. Cacheability
    - All REST response must be marked as cacheable or not cacheable.
4. Layered System
    - Client can not know or should not care whether it is connected directly to the server or to an intermediary.
5. Code on demand
    - Server are allowed to transfer executable code like JS and compiled components to clients.
6. Uniform Interface
    - Resource id in request: URI request must specify what resource it is looking for and what format the response should be.
    - Resource manipulation through representations: Once client has representation of resource, it can modify or delete the resource.
    - Self descriptive message: Each representation must describe its own data format
    - Hypermedia as the engine of application: Once client has access to REST service, it should be able to discover all available resource and methods through the hyperlink provided

* REST can work with HTTP, SMTP, FTP, etc.
* Meeting above constraint makes RESTful API.
